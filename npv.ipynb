### 4. Given an array with strings -  (5 Marks)

str_arr = np.array(['Python', 'Machine learning', 'Sql', 'Excel']).

Use inbuilt numpy functions to print the array in upper and lower case.

```python
str_arr = np.array(['Python', 'Machine learning', 'Sql', 'Excel'])

upper_list = []
lower_list= []

for lang in str_arr:
    upper_list.append(lang.upper())
    lower_list.append(lang.lower())
    
print('UPPER CASE array :')
print(upper_list)
print()
print('LOWER CASE array :')
print(lower_list)
```

D, Find the total no of suicides each year in descending order?

```python
suicides_each_year = pd.pivot_table(data = who_suicide_statistics, index= 'year', aggfunc = 'sum')
suicides_each_year.sort_values(by='suicides_no', ascending = False)['suicides_no']
```

```python
suicides_each_year.sort_values(by='suicides_no', ascending = False)['suicides_no'][:10].plot(kind='bar')
plt.show()
```

H.	Observe the column occupation, every entry has two parts: profession-specialty. Make two separate columns namely profession and specialty out of this column. Drop the column occupation from the dataset.

```python
Adult.head()
```

```python
profession_list = []
speciality_list = []
for occ in Adult['occupation']:
    words = occ.split('-')
    profession_list.append(words[0])
    speciality_list.append(words[1])
```

```python
Adult['profession'] = profession_list
Adult['speciality'] = speciality_list
```

```python
Adult.head()
```

```python
Adult.drop(columns=['occupation'], inplace= True)
```

```python
Adult.head(3)
```
